#Matplotlib

Matplotlib is the most commonly used Python plotting library. If you're coming from a Matlab background, the API is meant to emulate that style of workflow.

The Matplotlib + IPython stack is quite useful. As you'll see (through experimentation), matplotlib is complicated and large, and you will *usually* need some experimentation to figure out what to do.

```python
# -- set this so stuff happens in-notebook
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.power(x, 2)
plt.plot(x, y)
plt.title('An example plot!')
plt.show()
```

It's really easy! Packages exist that "wrap" graphics -- enter, seaborn. (Remember to `pip install seaborn`)

```python
import seaborn as sns
x = np.linspace(0, 10, 1000)
y = np.power(x, 2)
plt.plot(x, y)
plt.title('An example plot! Using Seaborn...')
plt.show()
```

Can get very involved!

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
f, ax = plt.subplots(1, 1, figsize=(5,4))
x = np.linspace(0, 10, 1000)
y = np.power(x, 2)
ax.plot(x, y)
ax.set_xlim((1, 5))
ax.set_ylim((0, 30))
ax.set_xlabel('my x label')
ax.set_ylabel('my y label')
# -- we can handle LaTeX!
ax.set_title(r'plot title, including $\int f d\Omega$')
plt.tight_layout()
# -- save to a PDF
plt.savefig('line_plot_plus.pdf')
```

Can histogram data! Very useful in the sciences.

```python
f, ax = plt.subplots(1, 1, figsize=(5,4))
bins = np.linspace(-20, 20)
plt.hist(np.random.normal(-1, 3, 1000), bins = bins, color='red', label = 'Trial', histtype='step')
plt.hist(np.random.normal(6, 4, 1000), bins = bins, color='blue', label = 'Control', histtype='step')
ax.set_xlabel(r'complicated label $\sum_i X_i$')
ax.set_ylabel('Count!')
plt.legend()
plt.show()
```

2D histogram! These can be pretty!

```python
f, ax = plt.subplots(1, 1, figsize=(5,4))
bins = np.linspace(-20, 20)
plt.hist2d(np.random.normal(-1, 3, 100000), np.random.normal(6, 4, 100000), bins=100)
ax.set_xlabel(r'complicated label $\sum_i X_i$')
ax.set_ylabel('Count!')
plt.show()
```
